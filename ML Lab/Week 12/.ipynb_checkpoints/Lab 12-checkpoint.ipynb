## The Normal Equation

### Direct computing to get the best parameters

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Creating synthetic dataset

```python
np.random.seed(42)
m = 100
X = 2 * np.random.rand(m, 1)
y = 4+3*X+np.random.randn(m, 1)
```

```python
X[:10], y[:10]
```

```python
plt.plot(X, y, "b.")
```

```python
from sklearn.preprocessing import add_dummy_feature
X_b = add_dummy_feature(X) # add x0 = 1 to each instance
theta_best = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
```

```python
theta_best
```

```python
# now predict the value of unseen data
test_data = np.array([[0], [2]])
# add bias
test_data_b = add_dummy_feature(test_data) # add x0 = 1 to each instance
y_predict = test_data_b @ theta_best
y_predict
```

```python
test_data
```

```python
# See on plot
plt.plot(X, y, "b.")
plt.plot(test_data, y_predict, "ro")
plt.plot(test_data, y_predict, "g--")
```

### Learing Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
m = 100 #dataset length
X = 2*np.random.rand(m, 1) # column vector
y = 4+3*X + np.random.rand(m, 1) # column vector
```

```python
from sklearn.preprocessing import add_dummy_feature
X_b = add_dummy_feature(X) # add x0 = 1 to each instance
```

```python
eta = 0.1 # Learning rate
n_epochs = 1000
m = len(X_b) # no of instances
np.random.seed(42)
theta = np.random.randn(2,1)
for epoch in range(n_epochs):
    gradients = 2/m * X_b.T @ (X_b @ theta - y)
    theta = theta - eta * gradients
```

```python
theta
```

```python
import matplotlib as mpl
from sklearn.metrics import mean_squared_error
X_new = np.array([[0], [2]])
X_new_b = add_dummy_feature(X_new) #add x0 = 1 to each instance

def plot_gradient_descent(theta, eta):
    m = len(X_b)
    plt.plot(X, y, "b.")
    n_epochs = 1000
    n_shown = 20
    theta_path = []
    loss = []
    for epoch in range(n_epochs):
        if epoch < n_shown:
            y_predict = X_new_b @ theta
            plt.plot(X_new, y_predict)
        gradients = 2/m*X_b.T @ (X_b @ theta - y)
        theta = theta - eta * gradients
        loss.append((1/m)*np.sum((X_b @ theta - y)**2))
        theta_path.append(theta)
    plt.xlabel("$x_1$")
    plt.axis([0,2,0,15])
    plt.grid()
    plt.title(fr"$\eta = {eta}$")
    return theta_path, loss
np.random.seed(42)
theta = np.random.randn(2, 1) #random weight initialization
theta_path_bgd, loss = plot_gradient_descent(theta, eta=0.02)
plt.ylabel("$y$", rotation=0)
```

```python
# plot the loss
plt.plot(loss[:20], 'r*')
plt.xlabel("Epochs")
plt.ylabel('loss')
plt.axis([0,20,0,40])
plt.title(fr"$\eta = {0.02}$")
```

```python
r, loss = plot_gradient_descent(theta, eta=0.1)
plt.show()
```
