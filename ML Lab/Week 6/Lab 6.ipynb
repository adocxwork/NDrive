```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#line plot
plt.plot([1,2,3,4,5], [0.5,0.75,1.25,1.7,2])
```

```python
# give name to x-axis and y-axis
plt.plot([1,2,3,4,5], [0.5,0.75,1.25,1.7,2])
plt.xlabel('X-Axis Data')
plt.ylabel('Y-Axis Data')
plt.title("Plot with text on both axis")
```

### Scatter Plot

```python
# give name to x-axis and y-axis
plt.scatter([1,2,3,4,5], [0.5,0.75,1.25,1.7,2], marker='^', c='red')
plt.xlabel('X-Axis Data')
plt.ylabel('Y-Axis Data')
plt.title("Plot with text on both axis")
```

```python
# give name to x-axis and y-axis
plt.scatter([1,2,3,4,5], [0.5,0.75,1.25,1.7,2], marker='^', c='red')
plt.plot([1,2,3,4,5], [0.5,0.75,1.25,1.7,2])
plt.xlabel('X-Axis Data')
plt.ylabel('Y-Axis Data')
plt.title("Plot with text on both axis")
```
