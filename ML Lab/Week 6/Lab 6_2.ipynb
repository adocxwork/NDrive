## Histogram in Matplotlib library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.hist([12,8,15,8,5,20], color='g')
```

```python
x=np.random.randint(10,60,(50))
```

```python
plt.hist(x, color='g', edgecolor='r')
plt.title("Histogram Demo")
plt.xlabel('City Code')
plt.ylabel('Population')
```

```python
plt.hist(x, color='g', bins=[10,20,30,40,50,60], edgecolor='r')
```

```python
len(x[x<20]), len(x[(x>=20) & (x<30)])
```

```python
# assign range
plt.hist(x, range=(0,100), edgecolor='r', color='g')
```

```python
plt.hist(x, "auto", edgecolor='r', color='g')
```

```python
plt.hist(x, "auto", edgecolor='r', color='g', cumulative=-1)
```

```python
# bottom parameter to change y-axis starting position
plt.hist(x, "auto", edgecolor='r', color='g', bottom=20)
```

```python
plt.hist(x, "auto", edgecolor='r', color='g', bottom=20, align='mid') #align=mid,left,right
```

```python
plt.hist(x, "auto", edgecolor='r', color='g', bottom=20, align='mid', histtype='stepfilled') #histtype = step,stepfilled,barstacked, bar
```

```python
plt.hist(x, "auto", edgecolor='r', color='g', orientation='horizontal') #horizontal,vertical
```

```python
plt.hist(x, "auto", edgecolor='r', color='g', rwidth=0.7) #space b/w towers
```

```python
plt.hist(x, "auto", edgecolor='r', color='g', rwidth=0.7, log=True) #y-axis values in log table
```

```python
plt.hist(x, "auto", edgecolor='r', color='g', rwidth=0.7, label="City-wise population")
plt.legend()
```

```python
#put a vertical line to partition the data add option axvline at x-axis value, say we want line at 35, redline
plt.hist(x, "auto", color='g', label="City-wise Population")
plt.axvline(35, color='r', label="break point")
plt.legend()
```
